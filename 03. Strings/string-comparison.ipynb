# 1. String Comparison

### Lexicographically

```python
str1 = 'banana'
str2 = 'banned'
```

```python
ord('a'), ord('n')
```

```python
str2 < str1
```

```python
str3 = 'ball'
str4 = 'balls'

str3 < str4
```

```python
'apple' == 'apple'
```

Case-sensitivity

```python
str1 = 'zoo'
str2 ='zoO'
```

```python
str1 == str2
```

```python
for ch1, ch2 in zip(str1, str2):
    print(ch1, ord(ch1), '-', ch2, ord(ch2))
```

```python
str1 > str2
```

### Sorting

```python
str1 = 'abcABC'
str1
```
