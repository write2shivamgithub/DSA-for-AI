# Lists

## 1D Operations

```python
### creation
numbers = [10, 20, 30, 40, 50]
fruits = ['apple', 'banana', 'cherry', 'date', 'elderberry']
mixed_list = [1, 'two', 3.0, 'four', 5]
```

```python
type(mixed_list)
```

```python
### access
numbers[3]
```

```python
fruits[1]
```

```python
### traversal
for fruit in fruits:
    print(fruit, end=' ')
```

### Insertion

```python
### append
numbers.append(60)
print(numbers)
```

```python
### insert
numbers.insert(2, 25) # insert 25 at index 2
```

### Merging

```python
### extend
more_fruits = ['fig', 'grape']
fruits.extend(more_fruits)
print(fruits)
```

```python
some_more_fruits = ('honeydew', 'kiwi')  # tuple also works
fruits.extend(some_more_fruits)
print(fruits)
```

```python
### concatenation
li1 = [1, 2, 3]; li2 = [4, 5, 6]
li3 = li1 + li2
print(li3)
```

### Remove

```python
### deletion
li = [3, 4, 6, 4, 1, 6, 2]
li.remove(6)  # removes first occurrence
print(li)
```

```python
### pop
li.pop()
print(li)   # removes last
```

```python
li.pop(2)
print(li)       # removes index 2 
```

```python
### delete
del li[1]  # deletes index 1
print(li)
```

```python
### clear
li.clear()
print(li, type(li))
```

```python
### updation
mixed_list[1] = 'TWO'
print(mixed_list)
```

```python
### count
li = [1, 2, 3, 2, 4, 2, 5]  
li.count(2)  # counts occurrences of 2
```

```python
### sort
li.sort() # in-place sort
print(li) 
```

```python
li1 = [3, 1, 4, 2]
sorted(li1) # returns new list
```

```python
### reverse
li1.reverse() # in-place reverse
print(li1)
```

```python
### copy
li2 = li1.copy() # independent copy
li2.append(5)
li2.pop(0)

print("li1:", li1)
print("li2:", li2) 
```

```python
### split(slicing)
print(numbers[:2]) # first half
print(numbers[1:]) # from index 1 to end
```

## 2D Operations

```python
### creation
matrix = [
    [1, 2, 3], 
    [4, 5, 6],
    [7, 8, 9]
    ]
```

```python
### display
def display_matrix(mat):
    for row in mat:
        print(row)
display_matrix(matrix)
```

```python
### traverse
for row in matrix:
    for ele in row:
        print(ele)
```

```python
### traverse
for row in matrix:
    for ele in row:
        print(ele, end = " ")
```

```python
### traverse
for row in matrix:
    for ele in row:
        print(ele, end = " ")
    print()  # New line after each row
```

```python
print(matrix[1][2])  # Accessing element at row 1, column 2 (value 6    )
```

```python
### updation
matrix[0][1] = 20  # Update element at row 0, column 1 to 20
display_matrix(matrix)
```

```python
### insertion
matrix[2].insert(1, 15)
display_matrix(matrix)
```

```python
matrix[0].append(20)
display_matrix(matrix)
```

```python
### deletion
matrix[0].pop()
matrix[2].remove(15)
display_matrix(matrix)
```
