### 1. Find the Largest Number in a List

```python
numbers = [12, 45, 7, 89, 34, 95, 62, 57]

# write code here
largest = numbers[0]
for element in numbers:
    if element > largest:
        largest = element
print(f'Largest element: {largest}')
```

### 2. Count Occurrences of an Element

```python
items = ['apple', 'banana', 'cherry','cherry', 'date', 'elderberry']
target = 'cherry'

# write code here
count = 0
for item in items:
    if item == target:
        count += 1
print(f'Count of "{target}": {count}')
```

### 3. Remove Duplicates from a List

```python
numbers  = [1, 2, 2, 3, 4, 4, 5]

# write code here
unique_elements = []
for number in numbers:
    if number not in unique_elements:
        unique_elements.append(number)
print(f'Unique elements: {unique_elements}')
```
