### 1. Array module

```python
import array
```

```python
### creation
arr1 = array.array('i', [10, 20, 30, 20, 50])
print(arr1)
```

```python
arr2 = array.array('i', [60, 70, 80])
print(arr2)
```

```python
### access
arr1[2]
```

```python
arr1[3]
```

```python
### traversal
for element in arr1:
    print(element)
```

```python
for element in arr1:
    print(element, end= " ")
```

```python
### insertion
arr1.insert(3, 35)
print(arr1)
```

```python
### deletion
arr1.remove(30)
print(arr1)
```

```python
### updation
arr1[2] = 30
print(arr1)
```

```python
### sorting
arr3 = array.array('i', [4, 7, 3, 9, 2, 6])
sorted(arr3)
```

```python
### merging
arr4 = arr1 + arr2
print(arr4)
```

```python
 ### splitting
 ix = int(len(arr4)/ 2)
 print(arr4[:ix])
print(arr4[ix:])
```

## 2. Numpy

```python
import numpy as np
```

```python
### creation
np1 = np.array([10, 20, 30, 20, 50])
np2 = np.array([60, 70, 80])
```

```python
### access
np1[0]
```

```python
### traversal
for ele in np1:
    print(ele)
```

```python
### insertion
np.insert(np1, 3, 35)
```

```python
### deletion
np.delete(np1, 3)
```

```python
### updation
np1[2] = 35
print(np1)
```

```python
### sorting
np3 = np.array([5, 8, 4, 9, 1, 6])
np.sort(np3)
```

```python
### merging
np3 = np.concatenate((np1, np2))
print(np3)
```

```python
### splitting
split1, split2 = np.array_split(np3, 2)
print(split1, split2)
```

### Vectorization

```python
np1 + 10 # add scalar to every element
```

```python
np1 * 2 # multiply every element
```

```python
np3 = np.array([4, 5])
np4 = np.array([6, 7])
np3 / np4 # element-wise division
```
